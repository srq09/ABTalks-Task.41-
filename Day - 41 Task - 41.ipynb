```python
"LangChain is a framework for developing applications powered by language models."

"Retrieval-Augmented Generation (RAG) combines retrieval systems with Large Language Models."

"FAISS is a vector database library developed by Meta for efficient similarity search."

"Embeddings convert text into numerical vector representations."

"Document chunking improves retrieval quality by splitting large documents into smaller pieces."
```

```python
# ==========================================
# DAY 41 - SIMPLE RAG PIPELINE DEMO
# ==========================================

documents = [
    "LangChain is a framework for building applications using Large Language Models.",
    "RAG stands for Retrieval Augmented Generation and combines retrieval with LLMs.",
    "FAISS is a vector search library developed by Meta.",
    "Embeddings convert text into numerical vectors for semantic search."
]

# ==========================================
# DOCUMENT CHUNKING
# ==========================================

chunks = documents

print("=" * 50)
print("DOCUMENT CHUNKING")
print("=" * 50)

for i, chunk in enumerate(chunks, start=1):
    print(f"Chunk {i}: {chunk}")

# ==========================================
# EMBEDDINGS (SIMULATED)
# ==========================================

print("\nEmbeddings Created Successfully")

# ==========================================
# VECTOR SEARCH (SIMULATED)
# ==========================================

def retrieve(query):

    query = query.lower()

    if "langchain" in query:
        return documents[0]

    elif "rag" in query:
        return documents[1]

    elif "faiss" in query:
        return documents[2]

    elif "embedding" in query:
        return documents[3]

    return "No relevant document found."


# ==========================================
# PROMPT ENGINEERING
# ==========================================

def generate_answer(question):

    context = retrieve(question)

    response = f"""
Question:
{question}

Retrieved Context:
{context}

Generated Answer:
{context}
"""

    return response


# ==========================================
# TEST QUERIES
# ==========================================

questions = [
    "What is LangChain?",
    "What is RAG?",
    "What is FAISS?",
    "What are Embeddings?"
]

print("\n")
print("=" * 50)
print("RAG RESULTS")
print("=" * 50)

for q in questions:
    print(generate_answer(q))
    print("-" * 50)
```
